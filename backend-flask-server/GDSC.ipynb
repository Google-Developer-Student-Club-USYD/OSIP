```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split


class NBClassify():

    # Bayes Theorem is P(y|X) = P(X|y) * P(y) / P(X)
    
    def classify(self, variables, outcome):
        # outcome => outcomes being trained on

        self.classes = np.unique(outcome)        
        self.count = len(self.classes)
        self.variable_count,self.rows = variables.shape 
        self.mean = variables.groupby(outcome).mean().values
        self.var = variables.groupby(outcome).var().values
        self.prior = (variables.groupby(outcome).apply(lambda x: len(x)) / self.rows).values

    def gaussian_density(self, ind, x):     
        # (1/√2pi*σ) * exp((-1/2)*((x-μ)^2)/(2*σ²))
        return np.exp(-((x-self.mean[ind])**2 / (2 * self.var[ind]))) / np.sqrt(2 * np.pi * self.var[ind]) 

    def posterior_p(self, x):

        posteriors = []

        # Posterior for Yes and No
        i = 0
        while i < self.count:
            posterior = np.log(self.prior[i])  + np.sum(np.log(self.gaussian_density(i, x)))
            posteriors.append(posterior)
            i+=1

        # Find maximum probabilty between classes for each row. If same then choose yes
        return self.classes[np.argmax(posteriors)]
        
    def predict(self, variables):
        predictions = []
        for variable in variables.values:
            predictions.append(self.posterior_p(variable))

        return predictions

def classify_nb(data):

    

    train, test = train_test_split(data, test_size=0.2)
    X_train, Y_train = train.iloc[:, :-1], train.iloc[:, -1]
    X_test, Y_test = test.iloc[:, :-1], test.iloc[:, -1]
    classifier = NBClassify()

    classifier.classify(X_train,Y_train)
    predictions = classifier.predict(X_test)

    return predictions

classify_nb(data)
```
